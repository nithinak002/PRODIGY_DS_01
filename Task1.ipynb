```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('WorldBank Data.xlsx')
df.head()
```

```python
df.tail()
```

droping rows with empty columns

```python
df = df.dropna()
df.head()
```

```python
df.shape
```

```python
df.describe()
```

internet usage:

```python

# Group the data by 'Region' and calculate Internet usage
internet_usage_by_region = df.groupby('Region')['Individuals using the Internet (% of population)'].max().reset_index()

#bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Individuals using the Internet (% of population)', data=internet_usage_by_region, palette='viridis')
plt.title('Internet Usage Across Regions')
plt.ylabel('Internet Usage (% of Population)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

```python
life_expectancy_by_year = df.groupby('Year')['Individuals using the Internet (% of population)'].max().reset_index()

# line chart
plt.figure(figsize=(20, 6))
plt.plot(life_expectancy_by_year['Year'], life_expectancy_by_year['Individuals using the Internet (% of population)'], marker='o', linestyle='-')

plt.title('Individuals using the Internet (% of population) per Year')
plt.xlabel('Year')
plt.ylabel('Individuals using the Internet (% of population)')
plt.grid(True)

plt.show()
```

life expectancy

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Life expectancy at birth (years)'], bins=20, kde=True, color='red')
plt.title('Distribution of Life Expectancy at Birth')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')

plt.show()
```

```python
life_expectancy_by_year = df.groupby('Year')['Life expectancy at birth (years)'].max().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(life_expectancy_by_year['Year'], life_expectancy_by_year['Life expectancy at birth (years)'], marker='o', linestyle='-')

plt.title('Life Expectancy Over the Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.grid(True)

plt.show()
```

GDP in USD

```python

# Group the data by 'Region' and calculate the max GPD
internet_usage_by_region = df.groupby('Region')['GDP (USD)'].max().reset_index()

# bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='GDP (USD)', data=internet_usage_by_region, palette='viridis')
plt.title('GDP (USD) Usage Across Regions')
plt.ylabel('GDP (USD)')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

electricity consumption

```python
electric_power_by_year = df.groupby('Year')['Electric power consumption (kWh per capita)'].max().reset_index()

# line chart
plt.figure(figsize=(12, 6))
plt.plot(electric_power_by_year['Year'], electric_power_by_year['Electric power consumption (kWh per capita)'], marker='o', linestyle='-')

plt.title('Electric Power Consumption Over the Years')
plt.xlabel('Year')
plt.ylabel('Electric Power Consumption (kWh per capita)')
plt.grid(True)

plt.show()
```
